## 导入工具包

```python
from cvs import *

import matplotlib.pyplot as plt
%matplotlib inline
```

## 初始化摄像头

```python
# 摄像头ID 0-后置 1-前置
Camera_ID = 0
```

```python
cap = cvs.VideoCapture(Camera_ID)
```

## 调用手机摄像头拍摄一帧画面

```python
img_bgr = cap.read()
```

```python
img_bgr.shape
```

```python
# 在手机屏幕上显示画面
cvs.imshow(img_bgr)
```

```python
# 在单元格里显示画面
plt.imshow(img_bgr[:,:,::-1])
plt.show()
```
